2. Logistic Classification

```
X = tf.constant([[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]], dtype=tf.float32, shape=[6, 2])    
Y = tf.constant([[0], [0], [0], [1], [1], [1]], dtype=tf.float32, shape=[6, 1])    

W = tf.Variable(tf.random.uniform(shape=[2, 1], name='weight', minval=0, maxval=1, dtype=tf.float32, seed=3))
b = tf.Variable(tf.random.uniform(shape=[1], name='bias', minval=0, maxval=1, dtype=tf.float32, seed=3))

@tf.function
def hypothesis(x, w, b):
    return tf.math.sigmoid(tf.linalg.matmul(x, w) + b)

@tf.function
def cost(x, w, b, y):
    H = hypothesis(x, w, b)
    return -tf.reduce_mean(y * tf.math.log(H) + (1 - y) * tf.math.log(1 - H))

optimizer = tf.keras.optimizers.SGD(learning_rate=1e-2)
loss = lambda: cost(X, W, b, Y) 

for step in range(20001):
    optimizer.minimize(loss, var_list=[W, b])
    if step % 500 == 0:
        print(step, cost(X, W, b, Y))

@tf.function
def predicted(x, w, b):
    return tf.cast(hypothesis(x, w, b) > 0.5, dtype=tf.float32)

@tf.function
def accuracy(x, w, b, y):
    return tf.reduce_mean(tf.cast(tf.equal(predicted(x, w, b), y), dtype=tf.float32))
    
print("\nHypothesis: {0}, \nCorrect (Y) {1} \nAccuracy {2}"\
      .format(hypothesis(X, W, b), predicted(X, W, b), accuracy(X, W, b, Y)))
```
