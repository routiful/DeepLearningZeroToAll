**2. Multivariable hypothesis**

```
x1_data = tf.constant([73., 93., 89., 96., 73.])    
x2_data = tf.constant([80., 88., 91., 98., 66.])    
x3_data = tf.constant([75., 93., 90., 100., 70.])    
y_data = tf.constant([152., 185., 180., 196., 142.])

w1 = tf.Variable(tf.random.uniform(shape=[], minval=0, maxval=1, dtype=tf.float32, seed=3))
w2 = tf.Variable(tf.random.uniform(shape=[], minval=0, maxval=1, dtype=tf.float32, seed=3))
w3 = tf.Variable(tf.random.uniform(shape=[], minval=0, maxval=1, dtype=tf.float32, seed=3))
b = tf.Variable(tf.random.uniform(shape=[], minval=0, maxval=1, dtype=tf.float32, seed=3))
print(x1_data.shape)
print(w1.shape)
print(b.shape)
@tf.function
def hypothesis(x1, x2, x3):
    return x1 * w1 + x2 * w2 + x3 * w3 + b

@tf.function
def cost(x1, x2, x3, y):
    return tf.reduce_mean(tf.square(hypothesis(x1, x2, x3) - y))

optimizer = tf.keras.optimizers.SGD(learning_rate=1e-5)
loss = lambda: cost(x1_data, x2_data, x3_data, y_data) 

for step in range(2001):
    optimizer.minimize(loss, var_list=[w1, w2, w3, b])
    if step % 10 == 0: 
        print("\tstep[{}] cost([W], b): {} y: {}".format( \
            step, cost(x1_data, x2_data, x3_data, y_data).numpy(), hypothesis(x1_data, x2_data, x3_data).numpy()))
```

**3. Hypothesis using matrix**

```
x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
y_data = [[152.], [185.], [180.], [196.], [142.]]

X = tf.Variable(x_data, shape=[None, 3])
Y = tf.Variable(y_data, shape=[None, 1])

W = tf.Variable(tf.random.uniform(shape=[3, 1], minval=0, maxval=1, dtype=tf.float32, seed=3))
b = tf.Variable(tf.random.uniform(shape=[1], minval=0, maxval=1, dtype=tf.float32, seed=3))

@tf.function
def hypothesis(x):
    return tf.matmul(x, W) + b

@tf.function
def cost(x, y):
    return tf.reduce_mean(tf.square(hypothesis(x) - y))

optimizer = tf.keras.optimizers.SGD(learning_rate=1e-5)
loss = lambda: cost(X, Y) 

for step in range(2001):
    optimizer.minimize(loss, var_list=[W, b])
    if step % 10 == 0: 
        print("\tstep[{}] cost([W], b): {} Y: {}".format( \
            step, cost(X, Y).numpy(), hypothesis(X).numpy()))
```
