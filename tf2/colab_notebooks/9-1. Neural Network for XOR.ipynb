**2. Ready to Dataset**

```
x_train = tf.constant([[0, 0], [0, 1], [1, 0], [1, 1]])
y_train = tf.constant([[0], [1], [1], [0]])

x_test = x_train
y_test = y_train

print("Shape of input_data: {}".format(x_train.shape))
print("Shape of labeled_data: {}".format(y_train.shape))

print("Shape of input_test_data: {}".format(x_test.shape))
print("Shape of labeled_test_data: {}".format(y_test.shape))
```

**3. Neural Network**

```
nn_model = tf.keras.Sequential()
nn_model.add(tf.keras.layers.Dense(
    1,
    input_shape=(2, ),
    use_bias=True,
    kernel_initializer='glorot_uniform',
    bias_initializer='glorot_uniform',
    activation='sigmoid'))

nn_model.summary()
```

```
nn_model.compile(
    optimizer='SGD',
    loss='mse',
    metrics=['mse']
)
```

```
history = nn_model.fit(
    x=x_train,
    y=y_train,
    epochs=1000,
    verbose=0
)
```

```
test_loss, test_acc = nn_model.evaluate(x_test, y_test, verbose=2)
nn_model_predictions = nn_model.predict(x_test)
print("pred: {}, actual: {}".format(nn_model_predictions, y_test))
```

**4. Deep Learning**

```
dl_model = tf.keras.Sequential()
dl_model.add(tf.keras.layers.Dense(
    2,
    input_shape=(2, ),
    use_bias=True,
    kernel_initializer='random_uniform',
    bias_initializer='random_uniform',
    activation='sigmoid'))
dl_model.add(tf.keras.layers.Dense(
    1,
    use_bias=True,
    kernel_initializer='random_uniform',
    bias_initializer='random_uniform',
    activation='sigmoid'))

dl_model.compile(
    optimizer=tf.optimizers.Adam(lr=0.01), # learning rate!!!!!
    loss='binary_crossentropy',
    metrics=['accuracy']
)

dl_model.summary()

history = dl_model.fit(
    x=x_train,
    y=y_train,
    epochs=10000,
    verbose=0
)

test_loss, test_acc = dl_model.evaluate(x_test, y_test, verbose=2)

dl_model_predictions = dl_model.predict(x_test)
print("pred: {}, actual: {}".format(dl_model_predictions, y_test))
```
